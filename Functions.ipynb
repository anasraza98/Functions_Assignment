5.  What are iterators in Python and how do they differ from iterables?

  - Iterable: An object that can be looped over like a list, Tuple, String.
  - Iterator: An object that remembers its position while iterating. It generates elements one at a time.

  Example: numbers = [1, 2, 3]  #Iterable
           it = iter(numbers)  #creating an iterator

           print(next(it))  #output: 1

6. Explain the concept of generators in Python and how they are defined.

  - Generators are functions that use yield instead of return. They generate values on the fly and are memory efficient.
  Example: def count_up():
               for i in range(3):
                   yield i
           gen = count_up()
           print(next(gen))   #output 1
           print(next(gen))   #output 2    

7. What are the advantages of using generators over regular functions?

  - Advantages are:
  * Memory Efficient - They don't store all values at once.
  * Faster Execution - Values are produced only when needed.
  * Great for large datasets - Can be used for reading big files.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

  - map() applies a function to all elements.
  Example: map(lambda x: x*x, [1, 2, 3])
           [1, 4, 9]

  - reduce() applies a function to combine elements.
  Example: reduce(lambda x, y: x+y, [1, 2, 3])
           6

  - filter() filters elements based on condition.
  Example: filter(lambda x: x>2, [1, 2, 3])
           [3]

```python
# PRACTICAL QUESTIONS :


#Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_even_numbers(numbers):
  return sum(n for n in numbers if n % 2 == 0)

print(sum_even_numbers([1, 2, 3, 4, 5, 6]))
```

```python
# Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
  if n < 2:
    return False
  for i in range(2, int(n**0.5) + 1):
    if n % i == 0:
      return False
  return True
print (is_prime(11))
print (is_prime(15))
```

```python
# Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
  def __init__(self,n):
    self.n = n
    self.a, self.b, self.count = 0, 1, 0

  def __iter__(self):
    return self

  def __next__(self):
    if self.count >= self.n:
      raise StopIteration
    self.a, self.b = self.b, self.a + self.b
    self.count += 1
    return self.a

G = FibonacciIterator(5)
print(list(G))


```

```python
#Write a generator function in Python that yields the powers of 2 up to a given exponent.

def power_of_two(n):
  for i in range(n):
    yield 2**i

G = power_of_two(5)
print(list(G))
```

```python
#Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_lyl(filepath):
  with open(filepath, "r") as file:
    for line in file:
      yield line.strip()

for line in read_file_lyl('file.txt'):
  print(line)


```

```python
#Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.


def C_to_F (temperatures): return list(map(lambda a: (a*9/5) + 32, temperatures))
print(C_to_F([0, 100, -30]))
```

```python
# Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(s): return ''.join(filter(lambda x: x.lower() not in 'aeiou', s))
print(remove_vowels("Hello World"))
```
